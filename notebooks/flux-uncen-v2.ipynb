### Loading from Huggingface

```python
import gc
import torch

gc.collect()
torch.cuda.empty_cache()
```

```python
from diffusers import AutoPipelineForText2Image
import torch

# Load the base model
pipeline = AutoPipelineForText2Image.from_pretrained("black-forest-labs/FLUX.1-dev", torch_dtype=torch.bfloat16).to('cuda')

# Load the uncensored LoRA weights
pipeline.load_lora_weights('enhanceaiteam/Flux-uncensored-v2', weight_name='lora.safetensors')

prompt = '''
A strong and athletic  man standing on a basketball court, dressed in a sleek basketball outfit untucked and relaxed.
The court features polished wooden flooring, basketball hoops in the background, and markings for the free throw and three-point lines.
The lighting is bright and natural, with a sense of movement and energy in the atmosphere, as if he’s about to start a game.
'''

# Generate an image
image = pipeline(prompt).images[0]
image.save('uncensored_image_hf.png')
```
